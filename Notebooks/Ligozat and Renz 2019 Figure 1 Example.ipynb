# Ligozat & Renz 2019 Figure 1 Example

## Imports

```python
import qualreas as qr
import os
from IPython.display import Image
```

## Path to Network & Algebra

```python
qr_path = os.path.join(os.getenv('PYPROJ'), 'qualreas')

alg_dir = os.path.join(qr_path, "Algebras")

janhunen2019_example_fig_1_file = os.path.join(qr_path, "Networks", "janhunen2019_example_fig_1.json")
```

## Ligozat & Renz 2019 Figure 1 Example

```python
Image("Images/janhunen2019_example_fig_1.png", width=700, height=100)
```

### Instantiate the Constraint Network 

```python
janhunen2019_example_fig_1_net = qr.Network(algebra_path=alg_dir, json_file_name=janhunen2019_example_fig_1_file)

print(janhunen2019_example_fig_1_net)
```

### Summarize the Network

```python
janhunen2019_example_fig_1_net.summary()
```

### Perform Constraint Propagation

```python
janhunen2019_example_fig_1_net.propagate()
janhunen2019_example_fig_1_net.summary()
```

### Singleton Labelings

```python
janhunen2019_example_fig_1_singleton_labelings = janhunen2019_example_fig_1_net.consistent_singleton_labelings()
```

```python
print(f"There are {len(janhunen2019_example_fig_1_singleton_labelings)} possible singleton labelings, as shown below:")
```

```python
count = 1
for labeling in janhunen2019_example_fig_1_singleton_labelings:
    print("------------------------")
    print(f"   Labeling #{count}")
    print("------------------------")
    labeling.summary()
    count += 1
    print(" ")
```

The 2nd labeling, above, is the one shown in Figure 1 in the paper.
